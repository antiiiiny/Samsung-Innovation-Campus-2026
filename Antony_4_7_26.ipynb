```python
import pandas as pd

data = {
    "Name": ['Arun','Bala','Chitra','Divya'],
    "Marks": [80,75,90, 42]
}
df = pd.DataFrame(data)
df['Result']=df['Marks'].apply(lambda x:'Pass' if x>=50 else 'Fail')
print(df)
```

```python
import pandas as pd

data = {
    'Student': ['A','B','C'],
    'Present_Days': [80,75,90]
}
df = pd.DataFrame(data)
df['Attendance_%']=(df['Present_Days']/90)*100
print(df)
```

```python
import pandas as pd

data = {
    'Dept': ['CSE','CSE','ECE','ECE'],
    'Marks': [80,75,90,95]
}
df = pd.DataFrame(data)
result = df.groupby('Dept')['Marks'].mean()
print(result)
```

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['A','B','C'],
    'Marks': [85, np.nan,90]
}
df = pd.DataFrame(data)
df['Marks'].fillna(df['Marks'].mean(),inplace=True)
print(df)
```

```python
#@title Vector Operations

import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

print("Vector a:",a)
print("Vector b:",b)

print("Addition:",a+b)
print("Dot Product:",np.dot(a,b))
print("Cross Product:",np.cross(a,b))
```

```python
#@title Vector Visualisation

import matplotlib.pyplot as plt

plt.quiver(0, 2, 2, 3)
plt.quiver(0, 0, 4, 1)

plt.xlim(0, 5)
plt.ylim(0, 5)
plt.title("Vector Visualization")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid()
```

```python
#@title Probability Simulation
import random

trials=1000
heads = 0

for _ in range (trials):
  if random.choice(["H","T"])=="H":
    heads+=1
print("Probability of getting heads:", heads/trials)
```

```python
#@title Statistics Example
import numpy as np

data = [10,20,30,40,50]

print("Mean:",np.mean(data))
print("Median:",np.median(data))
print("Standard Deviation:",np.std(data))
```

```python
#@title Simple AI model

from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1],[2],[3],[4]])
Y = np.array([50,20,40,70])

model = LinearRegression()
model.fit(X,Y)

new_data = np.array([[8]])
prediction = model.predict(new_data)
print("Prediction for 5:",prediction)
```

```python
#@title Trignometric Functions
import math

angle = math.radians(30)
print("Sine of 30 degrees:",math.sin(angle))
print("Cosine of 30 degrees:",math.cos(angle))
print("Tangent of 30 degrees:",math.tan(angle))
```

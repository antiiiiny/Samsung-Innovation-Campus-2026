#PYTHON MATRIX OPERATIONS USING NUMPY

- Matrices created using numpy.array()
- dot() or @ operator for matrix multiplication
- Transpose .T

```python
#@title Matrix multiplication
import numpy as np

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
C = np.dot(A,B)
print(C)

#using @ operator
C = B@A
print(C)

#using matmul()
C = np.matmul(A,B)
print(C)

# * only brings element wise multiplication
C = A*B
print(C)
```

```python
#@title Transpose of Matrix
import numpy as np

A = np.array([[1,2],[3,4]])
print(A)
print(A.T)
print(np.transpose(A))
```

```python
#@title 3x3 matrix mult and transpose
import numpy as np

A = eval(input("Enter first matrix: "))
B = eval(input("Enter second matrix: "))

A = np.array(A)
B = np.array(B)
C = A@B

print(C)
print(C.T)
```

```python
#@title Square matrix
import numpy as np
n = int(input("Enter size of square matrix (n): "))
print("Enter elements row-wise: ")
elements=[]
for i in range(n):
  row = list(map(int,input().split()))
  elements.append(row)

A = np.array(elements)
print("Square Matrix: ",A)
```

```python
#@title Diagonal Matrix(User input)
import numpy as np

n = int(input("Enter size of diagonal matrix (n): "))
print("Enter diagonal elements: ")
elements = list(map(int,input().split()))
D = np.diag(elements)
print("Diagonal Matrix: \n" ,D)
```

```python
#@title Unit matrix
import numpy as np
n= map(int,(input("Enter size of unit matrix (n)"))
unit_matrix = np.eye(n)
print("Unit Matrix: \n", unit_matrix)
```

```python
#@title Symmetric Matrix

import numpy as np
n = int (input("Enter size of matrix(n): "))
print("Enter elements row-wise: ")
elements = []
for i in range (n):
  row = list(map(int, input().split()))
  elements.append(row)

matrix = np.array(elements)
print("Matrix: ", matrix)
if np.array_equal(matrix, matrix.T):
  print("Symmetric Matrix")
else:
  print("Not a Symmetric Matrix")
```

```python
import numpy as np
n = int (input("Enter size of matrix(n): "))
print("Enter elements row-wise: ")
elements = []
for i in range (n):
  row = list(map(int, input().split()))
  elements.append(row)

matrix = np.array(elements)
print(np.linalg.inv(matrix))
```

```python
import numpy as np
A = np.array([[1,-4],[-4,1]])
B = np.array([[2,-8],[-8,2]])
print("A+B= \n", A+B)
print("A-(1/2)B= \n", A-(1/2)*B)
print("5A= \n", 5*A)
```

```python
import numpy as np

A = np.array([[4,2],[1,3]])
eigen_values, eigen_vectors = np.linalg.eig(A)
print("Eigen Values: \n", eigen_values)
print("Eigen Vectors: \n", eigen_vectors)
```

```python
import sympy as sp

x,y = sp.symbols('x y')
f = sp.sin(x*y)+sp.cos(x+y)
print("fx = ",sp.diff(f,x))
print("fy = ",sp.diff(f,y))
print("fxx = ", sp.diff(f,x,x))
print("fxy = ", sp.diff(f,x,y))
print("fyy = ", sp.diff(f,y,y))
```

```python
df = pd.read_csv('data_sales.csv')
df.head()
```

```python
# Convert 'Date' column to datetime objects first
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Day'] = df['Date'].dt.day
```

```python
total_units_region = df.groupby('Region')['Units'].sum()
print(total_units_region)
```

```python
df['Revenue'] = df['Units'] * df['Unit Price']
product_wise_sales = df.groupby('Item')[['Units', 'Revenue']].sum()
display(product_wise_sales)
```

```python
df.head()
```

```python
sorted_revenue = df.sort_values(by = 'Revenue', ascending = False)
sorted_revenue.head()
```

```python
time_based = df.groupby('Date')['Revenue'].sum()
time_based.head()
```

```python
df['Quarter'] = df['Date'].dt.quarter

quarterly_revenue = df.groupby('Quarter')['Revenue'].sum()

print(quarterly_revenue)
```

```python
 #@title Arithmetic Sequenec progamming

a = float (input("Enter the first term (a): "))
d = float (input("Enter the common difference (d): "))
n = int (input("Enter the number of terms (n): "))

sequence=[]
for i in range (n):
  an = a + i*d
  sequence.append(an)
print("Arithmetic Sequence: ", sequence)
nth = a + (n-1) *d
sum_terms = (n/2)*(2*a + (n-1)*d)
print("Nth Term: " , nth)
print("Sum of all the terms: ", sum_terms)
```

```python
 #@title Geometric Sequenece progamming

a = float (input("Enter the first term (a): "))
r = float (input("Enter the common ratio (r): "))
n = int (input("Enter the number of terms (n): "))

sequence=[]
for i in range (n):
  an = a * (r**i)
  sequence.append(an)
print("Geometric Sequence: ", sequence)
nth = a * (r**n)
if r==1:
  sum_terms = a*n
else:
  sum_terms = (a*(1-r**n))/(1-r)
print("Nth Term: " , nth)
print("Sum of all the terms: ", sum_terms)
```

```python
import math
import random
import matplotlib.pyplot as plt

# Generate realistic stock prices
def generate_stock_prices(initial_price, days):
    prices = [initial_price]

    for i in range(1, days):
        change_percent = random.uniform(-2, 2)  # random fluctuation
        new_price = prices[-1] * (1 + change_percent/100)
        prices.append(round(new_price, 2))

    return prices

# Euclidean Distance
def euclidean_distance(seq1, seq2):
    return math.sqrt(sum((a - b) ** 2 for a, b in zip(seq1, seq2)))

# Input
initial = float(input("Enter initial stock price: "))
days = int(input("Enter number of days: "))

# Generate stock
stock1 = generate_stock_prices(initial, days)

# Print values
print("\nStock Prices:")
for i, price in enumerate(stock1):
    print(f"Day {i+1}: {price}")

# Plot graph
plt.plot(range(1, days+1), stock1, marker='o', label="Stock 1")

# Optional comparison
choice = input("\nCompare with another stock? (yes/no): ")

if choice.lower() == "yes":
    initial2 = float(input("Enter second stock initial price: "))
    stock2 = generate_stock_prices(initial2, days)

    plt.plot(range(1, days+1), stock2, linestyle='--', marker='.', label="Stock 2")

    dist = euclidean_distance(stock1, stock2)
    print("\nEuclidean Distance:", round(dist, 2))

# Labels
plt.title("Realistic Stock Price Simulation")
plt.xlabel("Days")
plt.ylabel("Price")
plt.xlabel("Days", fontsize=12)
plt.ylabel("Price", fontsize=12)

# Grid styling
plt.grid(alpha=0.3)

# Legend
plt.legend()

# Remove top/right borders (clean look)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

# --- Analysis Section ---

initial1 = stock1[0]
final1 = stock1[-1]

initial2 = stock2[0]
final2 = stock2[-1]

# Percentage change
percent1 = ((final1 - initial1) / initial1) * 100
percent2 = ((final2 - initial2) / initial2) * 100

print("\n--- Stock Analysis ---\n")

print(f"Stock 1: Started at {initial1}, ended at {final1} ({percent1:.2f}% change)")
print(f"Stock 2: Started at {initial2}, ended at {final2} ({percent2:.2f}% change)\n")

# Better stock
if percent1 > percent2:
    print("👉 Stock 1 performed better overall.")
elif percent2 > percent1:
    print("👉 Stock 2 performed better overall.")
else:
    print("👉 Both stocks performed equally.")

# Volatility (simple check)
volatility1 = max(stock1) - min(stock1)
volatility2 = max(stock2) - min(stock2)

print("\nVolatility:")
print(f"Stock 1 fluctuation: {volatility1:.2f}")
print(f"Stock 2 fluctuation: {volatility2:.2f}")

if volatility1 > volatility2:
    print("👉 Stock 1 was more volatile (more risky).")
else:
    print("👉 Stock 2 was more volatile (more risky).")
```

```python
import random
import math
import matplotlib.pyplot as plt

# -----------------------------
# Generate stock data
# -----------------------------
def generate_stock(initial, days):
    prices = [initial]
    for _ in range(days - 1):
        change = random.uniform(-2, 2)
        new_price = prices[-1] * (1 + change / 100)
        prices.append(round(new_price, 2))
    return prices

# -----------------------------
# Create stocks
# -----------------------------
days = 120
stocks = {
    "A": generate_stock(200, days),
    "B": generate_stock(180, days),
    "C": generate_stock(220, days),
    "D": generate_stock(150, days)
}

# -----------------------------
# Show initial graph (NON-BLOCKING)
# -----------------------------
plt.figure(figsize=(10,5))
for name, data in stocks.items():
    plt.plot(data, label=f"Stock {name}")

plt.title("Stock Trends (120 Days)")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.grid(alpha=0.3)

plt.show(block=False)
plt.pause(5)
plt.close()

print("\nAvailable stocks: A, B, C, D\n")

# -----------------------------
# Safe input function
# -----------------------------
def get_stock_choice(prompt):
    while True:
        choice = input(prompt).upper()
        if choice in stocks:
            return choice
        else:
            print("❌ Invalid choice. Please enter A, B, C, or D.")

# -----------------------------
# User Inputs
# -----------------------------
s1 = get_stock_choice("Choose first stock (A/B/C/D): ")
s2 = get_stock_choice("Choose second stock (A/B/C/D): ")

shares = int(input("How many shares to buy: "))

print(f"\nYou can invest for up to {days} days.")
invest_days = int(input("How many days to invest: "))

# -----------------------------
# Validate days
# -----------------------------
if invest_days > days:
    print(f"Only {days} days available. Adjusted to {days}.")
    invest_days = days
elif invest_days <= 0:
    print("Invalid input. Using 1 day.")
    invest_days = 1

# -----------------------------
# Extended Graph (Past + Future)
# -----------------------------
plt.figure(figsize=(12, 6))

for name, data in stocks.items():
    past = data  # 120 days
    future = []

    # Generate future based on last value
    last_price = past[-1]

    for _ in range(invest_days):
        change = random.uniform(-2, 2)
        last_price = last_price * (1 + change / 100)
        future.append(round(last_price, 2))

    # X values
    past_days = list(range(len(past)))
    future_days = list(range(len(past), len(past) + invest_days))

    # Plot past
    plt.plot(past_days, past, linewidth=2, label=f"Stock {name} (Past)")

    # Plot future
    plt.plot(future_days, future, linestyle='--', linewidth=2, label=f"Stock {name} (Future)")

# Mark investment point
plt.axvline(x=119, linestyle=':', linewidth=2)
plt.text(119, plt.ylim()[1]*0.9, "Investment Point", rotation=90)

# Labels
plt.title("Stock Trends with Investment Projection")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.grid(alpha=0.3)

plt.show()

# -----------------------------
# Analyze ALL stocks
# -----------------------------
print("\n--- Overall Stock Analysis ---\n")

results = {}
for name, data in stocks.items():
    start = data[0]
    end = data[invest_days - 1]
    percent = ((end - start) / start) * 100
    results[name] = percent
    print(f"Stock {name}: {percent:.2f}% return")

best_stock = max(results, key=results.get)
worst_stock = min(results, key=results.get)

print(f"\n👉 Best performing stock: {best_stock}")
print(f"👉 Worst performing stock: {worst_stock}")

# -----------------------------
# Selected stock analysis
# -----------------------------
stock1 = stocks[s1]
stock2 = stocks[s2]

buy1 = stock1[0]
sell1 = stock1[invest_days - 1]

buy2 = stock2[0]
sell2 = stock2[invest_days - 1]

profit1 = (sell1 - buy1) * shares
profit2 = (sell2 - buy2) * shares

percent1 = ((sell1 - buy1) / buy1) * 100
percent2 = ((sell2 - buy2) / buy2) * 100

# Volatility
vol1 = max(stock1[:invest_days]) - min(stock1[:invest_days])
vol2 = max(stock2[:invest_days]) - min(stock2[:invest_days])

# Euclidean distance
dist = math.sqrt(sum((a - b) ** 2 for a, b in zip(stock1, stock2)))

# -----------------------------
# Final Output
# -----------------------------
print("\n--- Your Investment Analysis ---\n")

print(f"You chose Stock {s1} and Stock {s2}\n")

print(f"Stock {s1}: Profit = {profit1:.2f}, Return = {percent1:.2f}%")
print(f"Stock {s2}: Profit = {profit2:.2f}, Return = {percent2:.2f}%\n")

if profit1 > profit2:
    print(f"👉 Stock {s1} performed better.")
elif profit2 > profit1:
    print(f"👉 Stock {s2} performed better.")
else:
    print("👉 Both performed equally.")

print("\nVolatility:")
print(f"{s1}: {vol1:.2f}")
print(f"{s2}: {vol2:.2f}")

if best_stock in [s1, s2]:
    print(f"\n✅ Good choice! {best_stock} was the best overall performer.")
else:
    print(f"\n⚠️ Better option existed: Stock {best_stock} was best.")

print(f"\nEuclidean Distance between trends: {dist:.2f}")
```

```python
import yfinance as yf
import random
import math
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# ==============================
# STOCK LIST
# ==============================
symbols = {
    "TCS": "TCS.NS", "RELIANCE": "RELIANCE.NS", "INFOSYS": "INFY.NS",
    "HDFC": "HDFCBANK.NS", "ICICI": "ICICIBANK.NS",
    "SBI": "SBIN.NS", "ITC": "ITC.NS", "WIPRO": "WIPRO.NS",
    "LT": "LT.NS", "BHARTI": "BHARTIARTL.NS",
    "AXIS": "AXISBANK.NS", "KOTAK": "KOTAKBANK.NS",
    "ASIANPAINTS": "ASIANPAINT.NS", "MARUTI": "MARUTI.NS",
    "TITAN": "TITAN.NS",
    "SUNPHARMA": "SUNPHARMA.NS", "ULTRATECH": "ULTRACEMCO.NS",
    "NESTLE": "NESTLEIND.NS", "POWERGRID": "POWERGRID.NS",
    "NTPC": "NTPC.NS",
    "ONGC": "ONGC.NS", "COALINDIA": "COALINDIA.NS",
    "BAJAJFINANCE": "BAJFINANCE.NS", "ADANI": "ADANIENT.NS",
    "HINDUNILEVER": "HINDUNILVR.NS"
}

options = list(symbols.keys())

# ==============================
# STEP 1: SELECT 4 STOCKS
# ==============================
d1 = widgets.Dropdown(options=options, description='Stock 1:')
d2 = widgets.Dropdown(options=options, description='Stock 2:')
d3 = widgets.Dropdown(options=options, description='Stock 3:')
d4 = widgets.Dropdown(options=options, description='Stock 4:')

btn1 = widgets.Button(description="Load Stocks", button_style='info')

display(d1, d2, d3, d4, btn1)

# ==============================
# FUNCTIONS
# ==============================
def get_stock_data(symbol):
    stock = yf.Ticker(symbol)
    hist = stock.history(period="120d")
    return hist['Close'].tolist()

def simulate_future(data, days):
    future = []
    last_price = data[-1]
    for _ in range(days):
        change = random.uniform(-4, 4)
        last_price = last_price * (1 + change / 100)
        future.append(round(last_price, 2))
    return future

# ==============================
# STEP 1 HANDLER
# ==============================
def load_stocks(b):
    clear_output(wait=True)

    selected4 = [d1.value, d2.value, d3.value, d4.value]

    print("Fetching data...\n")

    stocks = {}
    for name in selected4:
        stocks[name] = get_stock_data(symbols[name])

    # -------- GRAPH 1 --------
    plt.figure(figsize=(10,5))
    for name, data in stocks.items():
        plt.plot(data, linewidth=2, label=name)

    plt.title("Stock Prices (Last 120 Days)")
    plt.xlabel("Days")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

    print("\nNow choose 2 stocks to invest:\n")

    # ==============================
    # STEP 2 UI (ONLY FROM SELECTED 4)
    # ==============================
    invest_options = selected4

    s1 = widgets.Dropdown(options=invest_options, description='Invest 1:')
    s2 = widgets.Dropdown(options=invest_options, description='Invest 2:')

    shares = widgets.IntText(value=100, description='Shares:')
    days = widgets.IntText(value=30, description='Future Days:')

    btn2 = widgets.Button(description="Run Investment Analysis", button_style='success')

    display(s1, s2, shares, days, btn2)

    # ==============================
    # STEP 2 HANDLER
    # ==============================
    def run_analysis(b2):
        clear_output(wait=True)

        display(s1, s2, shares, days, btn2)

        invest_days = days.value
        num_shares = shares.value

        # -------- GRAPH 2 --------
        plt.figure(figsize=(12,6))

        future_data = {}

        for name, data in stocks.items():
            future = simulate_future(data, invest_days)
            future_data[name] = future

            past_days = list(range(len(data)))
            future_days = list(range(len(data), len(data)+invest_days))

            plt.plot(past_days, data, linewidth=2, label=f"{name} (Past)")
            plt.plot(future_days, future, linestyle='--', linewidth=2, label=f"{name} (Future)")

        plt.axvline(x=119, linestyle=':', linewidth=2)
        plt.title("Stock Trends (Real + Future)")
        plt.xlabel("Days")
        plt.ylabel("Price")
        plt.legend()
        plt.grid(alpha=0.3)
        plt.show()

        # -------- OVERALL ANALYSIS --------
        print("\n--- Overall Stock Analysis ---\n")

        results = {}
        for name in stocks:
            start = stocks[name][-1]
            end = future_data[name][-1]
            percent = ((end - start) / start) * 100
            results[name] = percent
            print(f"{name}: {percent:.2f}% return")

        best = max(results, key=results.get)
        worst = min(results, key=results.get)

        print(f"\n👉 Best performing stock: {best}")
        print(f"👉 Worst performing stock: {worst}")

        # -------- INVESTMENT ANALYSIS (ONLY 2) --------
        print("\n--- Your Investment Analysis ---\n")

        selected = [s1.value, s2.value]

        # Show chosen stocks performance
        for name in selected:
            buy = stocks[name][-1]
            sell = future_data[name][-1]

            profit = (sell - buy) * num_shares
            percent = ((sell - buy) / buy) * 100

            print(f"{name}: Profit = {profit:.2f}, Return = {percent:.2f}%")

        # -----------------------------
        # Compare with ALL 4 stocks
        # -----------------------------
        print("\n--- Decision Evaluation ---\n")

        # Best overall stock
        best_stock = max(results, key=results.get)

        # Rank all stocks
        sorted_stocks = sorted(results.items(), key=lambda x: x[1], reverse=True)

        print("Stock Ranking:")
        for i, (name, val) in enumerate(sorted_stocks, 1):
            print(f"{i}. {name} ({val:.2f}%)")

        print(f"\n👉 Best possible stock: {best_stock}")

        # -----------------------------
        # Smart Feedback
        # -----------------------------
        if best_stock in selected:
            print(f"\n✅ Good decision! You selected {best_stock}, which was the best performer.")
        else:
            print(f"\n⚠️ You missed the best stock ({best_stock}).")

            # Suggest better choice
            print(f"👉 Better choice would have been: {best_stock}")

        # Compare between chosen 2
        if results[selected[0]] > results[selected[1]]:
            print(f"\n👉 Between your picks, {selected[0]} performed better.")
        elif results[selected[1]] > results[selected[0]]:
            print(f"\n👉 Between your picks, {selected[1]} performed better.")
        else:
            print("\n👉 Both your picks performed equally.")

        # -----------------------------
        # NORMALIZED EUCLIDEAN DISTANCE
        # -----------------------------
        def normalize(data):
            min_val = min(data)
            max_val = max(data)

            if max_val == min_val:
                return [0 for _ in data]  # avoid division by zero

            return [(x - min_val) / (max_val - min_val) for x in data]

        # Normalize selected stocks
        norm1 = normalize(stocks[s1.value])
        norm2 = normalize(stocks[s2.value])

        # Compute distance
        distance = math.sqrt(sum((a - b) ** 2 for a, b in zip(norm1, norm2)))

        # Max possible distance (worst case)
        max_distance = math.sqrt(len(norm1))  # since values in [0,1]

        # Convert to similarity score
        score = (1 - (distance / max_distance)) * 100

        # Clamp score between 0 and 100
        score = max(0, min(100, score))

        print(f"\nSimilarity Score between {s1.value} and {s2.value}: {score:.2f} / 100")

        # Interpretation
        if score > 80:
            print("👉 Very similar trends")
        elif score > 60:
            print("👉 Moderately similar trends")
        elif score > 40:
            print("👉 Some similarity")
        else:
            print("👉 Very different behavior")

    btn2.on_click(run_analysis)

btn1.on_click(load_stocks)
```

```python
#@title Composite functions

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 10 + x**2

def g(x):
  return 3+x

def fog(x):
  return f(g(x))

def gof(x):
  return g(f(x))

x = np.linspace(-5, 5, 100)

plt.plot (x, fog(x), label = "f(g(x)) = 10 + (3+x)^2")
plt.plot (x, gof(x), label = "g(f(x)) = 3+ (10 + x^2)")

plt.axhline(0)
plt.axvline(0)

plt.title("Composite functions")
plt.legend()
plt.grid()
plt.show()
```

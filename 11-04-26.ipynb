```python
import sympy as sp

x = sp.symbols('x')

def chain_rule_explain(expr):
    if isinstance(expr, sp.Function):
        outer = expr.func
        inner = expr.args[0]

        print(f"\nFunction: {expr}")
        print(f"Outer function: {outer}")
        print(f"Inner function: {inner}")

        u = sp.symbols('u')
        outer_expr = outer(u)
        d_outer = sp.diff(outer_expr, u)

        print(f"Derivative of outer (d/dx {outer}(u)): {d_outer}")

        d_inner = sp.diff(inner, x)
        print(f"Derivative of inner (d/dx {inner}): {d_inner}")

        final = d_outer.subs(u, inner) * d_inner

        print(f"Final derivative using chain rule: {final}")
        return final
    else:
        print("Not a composite function")
        return sp.diff(expr, x)


def differentiate(expr_str):
    expr = sp.sympify(expr_str)

    print("\n--- Applying Chain Rule ---")
    result = chain_rule_explain(expr)

    print("\nFinal Answer:", result)


while True:
    user_input = input("\nEnter function (or 'exit'): ")
    if user_input.lower() == 'exit':
        break

    differentiate(user_input)
```

```python
import sympy as sp
x, y = sp.symbols('x y')
f = sp.exp(x**2 + 2*y**2)

fx = sp.diff(f, x)
print(f"fx = {fx}")

fxx = sp.diff(fx, x)
print(f"fxx = {fxx}")

fxy = sp.diff(fx, y)
print(f"fxy = {fxy}")
```

##DataSet 2 :
Columns: StudyHours, Attendance, Marks, SleepHours


```python
import numpy as np

data = np.genfromtxt('Student.csv', delimiter = ',', skip_header = 1, dtype=str)
print(data)
```

```python
study_hours = data[::,0].astype(float)
attendance = data[::,1].astype(float)
marks = data[::,2].astype(float)
sleep_hours = data[::,3].astype(float)
print(study_hours)
print(attendance)
print(marks)
print(sleep_hours)
```

```python
mean = np.mean(marks)
print("Mean: ", mean)
median = np.median(marks)
print("Median: ", median)
std_dev = np.std(marks)
print("Std Dev: ", std_dev)
```

```python
cor_studyhours_marks = np.corrcoef(study_hours,marks)
print(cor_studyhours_marks)
cor_attendance_marks = np.corrcoef(attendance,marks)
print(cor_attendance_marks)
```

```python
norm_marks = (marks - np.mean(marks))/np.std(marks)
print(norm_marks)
```

```python
marks_per_studyhour = marks/study_hours
print(marks_per_studyhour)
```

```python
top_students = data[marks>90]
print(top_students)
```

##DataSet - 3
Columns: Size, Sqft, Bedrooms, Age, Price

```python
import numpy as np

data = np.genfromtxt("House.csv", delimiter = ',', skip_header = 1, dtype=str )
print(data)
```

```python
size = data[::,0].astype(int)
sqft = data[::,1].astype(float)
bedrooms = data[::,2].astype(int)
age = data[::,3].astype(int)
price = data[::,4].astype(float)

print("Size:", size)
print("Sqft:", sqft)
print("Bedrooms:", bedrooms)
print("Age:", age)
print("Price:", price)
```

```python
mean_price = np.mean(price)
median_price = np.median(price)
std_dev_price = np.std(price)

print("Mean Price:", mean_price)
print("Median Price:", median_price)
print("Standard Deviation Price:", std_dev_price)
```

```python
above_average_houses = data[price > mean_price]
print("Houses above average price:")
print(above_average_houses)
```

```python
corr_sqft_price = np.corrcoef(sqft, price)
print("Correlation between Square Footage and Price:")
print(corr_sqft_price)

corr_bedrooms_price = np.corrcoef(bedrooms, price)
print("Correlation between Bedrooms and Price:")
print(corr_bedrooms_price)
```

```python
norm_price = (price - mean_price) / std_dev_price
print("Normalized Price:")
print(norm_price)
```

```python
price_per_sqft = price / sqft
print("Price per Square Foot:")
print(price_per_sqft)
```

```python

expensive_houses = data[price > 600000]
print("Expensive houses (price > 600000):")
print(expensive_houses)
```
